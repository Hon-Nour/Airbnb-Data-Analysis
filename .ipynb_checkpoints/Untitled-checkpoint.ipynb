```python
# Importing necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
%matplotlib inline
```

```python
def convert_dtype(x):
    if not x:
        return ''
    try:
        return str(x)   
    except:        
        return ''
df_listings = pd.read_csv('United_Kingdom_listing.csv', encoding = 'ISO-8859-1',converters={'host_response_time': convert_dtype,'district': convert_dtype})
df_listings.head()
```

```python
print( f"df_listings has {df_listings.shape[0]} rows and {df_listings.shape[1]} columns\n")
```

```python
# Let's see the columns we have
print("df_listings columns are:")
df_listings.columns
```

```python
# Selecting necessary variables for further analysis
df_listings = df_listings[['listing_id', 'name', 'host_is_superhost', 'neighbourhood',
       'city', 'latitude', 'longitude', 'property_type',
       'room_type', 'accommodates', 'bedrooms', 'amenities', 'minimum_nights', 'maximum_nights', 'review_scores_accuracy', 'review_scores_communication',
       'review_scores_location', 'instant_bookable', 'price']]
df_listings.head()
```

```python
df_listings.info()
```

```python
# Filling null values with 0
df_listings.fillna(0, inplace=True)
```

```python
# Subset prices from df_listings into df_price
df_price = df_listings['price']
# Create an empty price list
prices=[]

# Convert prices from df_listings into float values and append it in prices list
for p in df_price:
    p = float(p)
    prices.append(p)
    
# Replace the price column in the original listingDF with the new prices
df_listings['price'] = prices
```

```python
# Taking out listings with no price and review acuracy
df_listings = df_listings[df_listings['price'] > 0]
df_listings = df_listings[df_listings['bedrooms'] > 0]
df_listings = df_listings[df_listings['review_scores_accuracy'] > 0]
df_listings = df_listings[df_listings['minimum_nights'] > 0]
df_listings = df_listings[df_listings['maximum_nights'] > 0]
# Take out duplicates  
df_listings.drop_duplicates(subset="listing_id")
```

```python
print( f"df_listings now has {df_listings.shape[0]} rows and {df_listings.shape[1]} columns\n")
```

ANALYZING THE LISTINGS CHARACTERISTICS

```python
# Neighbourhoods of our listing
print("Number of Neighbourhoods: ", len(df_listings['neighbourhood'].unique()))
print()
print()
print("They are: ")
print()
print(df_listings["neighbourhood"].value_counts())
print()
g = sns.catplot(x="neighbourhood", data= df_listings, kind = "count", height=6, aspect=3)
# Showing the first 100 neighbourhoods
plt.xlim(0,50)
plt.xticks(rotation=90)
plt.show()
```

```python
# Assign neighbourhood value_counts to a DataFrame
order = df_listings['neighbourhood'].value_counts().to_frame()
order = order.reset_index()
# change column names
order.columns = ['Neighbourhood', 'counts'] 
order.head()
```

```python
# Order Plot by order dataframe
# plot the top 100
g = sns.catplot(x="neighbourhood", data= df_listings, kind = "count", height=7, aspect=2, order=order['Neighbourhood'])
# Showing the top 50 neighbourhoods in terms of no of listings
plt.xlim(0,50)
plt.xticks(rotation=90)
plt.show()
```

```python
# property types of our listing
print("Number of Property types: ", len(df_listings['property_type'].unique()))
print()
print()
print("They are: ")
print('property Type              | listings')
print(df_listings['property_type'].value_counts())
print()
sns.catplot(x="property_type", data= df_listings, kind = "count",height=12, aspect=3)
plt.xticks(rotation=90)
plt.show()
```

```python
# Assign value_counts to a DataFrame
order = df_listings['property_type'].value_counts().to_frame()
order = order.reset_index()
# change column names
order.columns = ['Property', 'counts'] 
order.head()
```

```python
# Order Plot by order dataframe
# plot the top 100
g = sns.catplot(x='property_type', data= df_listings, kind = "count", height=7, aspect=2, order=order['Property'])
# Showing the first 100 neighbourhoods
plt.xlim(0,50)
plt.xticks(rotation=90)
plt.show()
```

```python
print("Number of Room types: ", len(df_listings['room_type'].unique()))
print()
print()
print("They are: ")
print()
print(df_listings['room_type'].value_counts())
print()
# sns.catplot(x="room_type", data= df_listings, kind = "count", height=7, aspect=2, palette="flare")
# sns.set_style("dark")
plt.show()
```

```python
# # Assign value_counts to a DataFrame
# rooms = df_listings['room_type'].value_counts().to_frame()
# rooms = rooms.reset_index()
# # change column names
# rooms.columns = ['Room', 'counts'] 
# rooms
rooms = df_listings.groupby('room_type')['room_type'].agg('count').sort_values(ascending = False)
rooms
```

```python
label = [i for i in rooms.index]

def func(x, rooms):
    #pct = (x*100)/room_stats.sum()
    return "{:.1f}%".format(x)
          
fig = plt.figure(figsize = (8,8))

plt.pie(rooms,
       labels=label,
       autopct = lambda pct: func(pct, rooms),
       explode = [0.2, 0.1, 0.1, 0.2],
       shadow = True,)
plt.style.use('tableau-colorblind10')
plt.title("Types of rooms")
plt.show()
```

```python
print("Number of Bedrooms: ", len(df_listings['bedrooms'].unique()))
print()
print()
print("They are: ")
print()
print(df_listings['bedrooms'].value_counts())
print()
sns.catplot(x="bedrooms", data= df_listings, kind = "count", height=7, aspect=2)
plt.xlim(0,30)
# sns.set_style("dark")
plt.xticks(rotation=45)
plt.figure(dpi=300)
plt.show()
```

## Analyzing Price on based Lististings Characteristics

```python
plt.figure(figsize = (20,18))
sns.heatmap(df_listings.groupby(['city','room_type']).price.mean().unstack(), annot=True, fmt=".0f", cmap = sns.cm.rocket_r)
```

```python
# Checking out the mean prices for the different room and property types
df_room_price = df_listings.groupby(['neighbourhood','room_type']).price.mean()
df_room_price = df_room_price.reset_index()
df_room_price = df_room_price.sort_values('price', ascending=False)
df_room_price = df_room_price.sort_values('room_type')
```

```python
df_neighbourhood_room_price  = df_room_price.pivot_table(index="neighbourhood", columns="room_type", values="price")

df_neighbourhood_room_price.fillna(0, inplace=True)
```
